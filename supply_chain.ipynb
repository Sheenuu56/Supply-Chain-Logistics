```python
#Filling the missing values
df['Logistics_Delay_Reason'].fillna('Unknown', inplace= True)
df.head()
```

```python
df.dtypes
```

```python
df['Timestamp']= pd.to_datetime(df['Timestamp'])
df.dtypes
```

```python
#Creating some new columns for analysis
df['Date']= df['Timestamp'].dt.date
df['Time']= df['Timestamp'].dt.time
df['Month']= df['Timestamp'].dt.month_name()
df['Weekday']= df['Timestamp'].dt.day_name()
df['Hour']= df['Timestamp'].dt.hour
df.head()
```

```python
#Checking for outliers
import seaborn as sns
sns.boxplot(x=df['Waiting_Time'])
```

```python
Q1 = df['Waiting_Time'].quantile(0.25)
Q3 = df['Waiting_Time'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
```

```python
#Identifying Outliers
outliers = df[(df['Waiting_Time'] < lower_bound) | (df['Waiting_Time'] > upper_bound)]
print(outliers)
```

```python
df.dtypes
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
categorical_columns
```

```python
exclude_columns= ['Date', 'Time', 'Month', 'Weekday']
categorical_columns= [col for col in df.select_dtypes(include=['object']).columns if col not in exclude_columns]

for col in categorical_columns:
    df[col] = df[col].astype('category').cat.codes
```
